```python
import matplotlib.pyplot as plt
from PIL import Image
import math
import os

DPI=150
fig_dir="figures/final"

def save_figs(fig_name: str, fig_dir=fig_dir):
    plt.savefig(os.path.join(fig_dir, "pdf", f"{fig_name}.pdf"), dpi=DPI, bbox_inches="tight")
    plt.savefig(os.path.join(fig_dir, "eps", f"{fig_name}.eps"), format="eps", bbox_inches="tight")

    

def open_eps(filename, dpi=DPI):
    img = Image.open(filename)
    original = [float(d) for d in img.size]
    scale = dpi/72.0            
    if dpi is not 0:
        img.load(scale = math.ceil(scale))
    if scale != 1:
        img.thumbnail([round(scale * d) for d in original], Image.ANTIALIAS)
    return img
```

#### Offline results: Instantaneous prediction versus target tendency profile at lon=0 deg transect

```python
a = open_eps('figures/eps/offline_dQ1_transect_lon0.eps', dpi=DPI)
b = open_eps('figures/eps/offline_dQ2_transect_lon0.eps', dpi=DPI)


f, axes = plt.subplots(2, 1, figsize=(10,8), squeeze=False)

ax=axes[0][0]
ax.imshow(a)
ax.axis("off")
ax.text(0, 1., "a)", va='top', fontsize=18, transform=ax.transAxes)

ax=axes[1][0]
ax.imshow(b)
ax.axis("off")
ax.text(0, 1., "b)", va='top', fontsize=18, transform=ax.transAxes)

plt.tight_layout()

save_figs(fig_name="offline_transect")
```
